```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('losses_rnn.npy', 'rb') as f:
    loss_values_rnn = np.load(f)
with open('losses_gru.npy', 'rb') as f:
    loss_values_gru = np.load(f)
with open('losses_lstm.npy', 'rb') as f:
    loss_values_lstm = np.load(f)
```

```python
loss_values_lstm
```

```python
epochs = range(1, len(loss_values_rnn)+1)

plt.plot(epochs, loss_values_rnn, label='RNN Loss')
plt.plot(epochs, loss_values_gru, label='GRU Loss')
plt.plot(epochs, loss_values_lstm, label='LSTM Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```
